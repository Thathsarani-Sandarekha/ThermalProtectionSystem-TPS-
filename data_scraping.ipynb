```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Base URL of the website
base_url = "https://thermtest.com/thermal-resources/materials-database"

# Initialize an empty DataFrame to store results
columns = ["Material", "Thermal Conductivity (W/m*K)", "Thermal Diffusivity (mm2/s)", 
           "Specific Heat Capacity (J/kg*K)", "Volumetric Specific Heat (MJ/m³K)", 
           "Thermal Effusivity (Ws0.5/(m2K))", "Material Density (kg/m3)"]
all_data = []
```

```python
# Function to scrape one page
def scrape_page(page):
    response = requests.get(base_url + f"?page={page}")
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Locate the table
    table = soup.find('table')
    rows = table.find_all('tr')[1:]  # Skip header row
    
    for row in rows:
        cells = row.find_all('td')
        if len(cells) > 0:
            data = [cell.text.strip() for cell in cells]
            all_data.append(data)
```

```python
# Loop through all pages
for page in range(1, 109):  # Assuming 108 pages
    print(f"Scraping page {page}...")
    scrape_page(page)
```

```python
# Save to CSV
df = pd.DataFrame(all_data, columns=columns)
df.to_csv("thermal_properties_data.csv", index=False)
print("Data extraction complete! File saved as 'thermal_properties_data.csv'.")
```

### Random Sampling

```python
import pandas as pd
import random

# Step 1: Load the CSV file
file_path = 'enhanced_materials_dataset.csv'  # Replace with the path to your CSV file
data = pd.read_csv(file_path)

# Step 2: Shuffle the rows
shuffled_data = data.sample(frac=1, random_state=42).reset_index(drop=True)  # Randomly shuffle the data

# Step 3: Split the data into 3 unique datasets
df_train = shuffled_data.iloc[:2000]  # First 2000 rows
df_ga = shuffled_data.iloc[2000:4000]  # Next 2000 rows
df_test = shuffled_data.iloc[4000:4400]  # Next 400 rows

# Check if there are enough rows in the original data
if len(shuffled_data) < 4400:
    raise ValueError("The original CSV file does not have enough rows to create three unique datasets (2000, 2000, 400).")

# Step 4: Save each dataset to a new CSV file
df_train.to_csv('df_train.csv', index=False)
df_ga.to_csv('df_ga.csv', index=False)
df_test.to_csv('df_test.csv', index=False)

print("Datasets have been created and saved as 'df_train.csv', 'df_ga.csv', and 'df_test.csv'.")
```
